```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./data/results.csv')
```

```python
data
```

```python
df = data.loc[:,['driverId','constructorId']] # 드라이버들 중 다른 회사로 옮긴 적이 있는 비율 찾기
df
```

```python
df_1 = df.groupby(['driverId','constructorId']).count() #드라이버별 그룹화 진행
df_1
```

```python
df_list = np.unique(df) # 중복제거
df_list
```

```python
df_1.index 
```

```python
df_1.index[0][0]
```

```python
df_1.index.shape
```

```python
# 2팀 이상 머무른 전적이 있는 선수들
driver_list = []
for i in df_list:
    j = 0
    for k in range(2126):
        if df_1.index[k][0] == i:
            j += 1

    if j > 1:
        driver_list.append(i)
```

```python
len(driver_list)
```

```python
# 3팀 이상 머무른 전적이 있는 선수들
driver_list_1 = []
for i in df_list:
    j = 0
    for k in range(2126):
        if df_1.index[k][0] == i:
            j += 1

    if j > 2:
        driver_list_1.append(i)
```
