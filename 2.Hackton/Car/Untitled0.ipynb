```python
from google.colab.patches import cv2_imshow
from PIL import Image
import cv2

url = '/content/drive/MyDrive/Personal_Training_To_Github/hackerton/02_(원본데이터)2_1_postoffice_20201222_141259 14.jpg'
image = cv2.imread(url)
#image = cv2.imread('opencv_data\\'+'')  
#cv2_imshow(image)

xml = cv2.CascadeClassifier('/drive/MyDrive/Personal_Training_To_Github/hackerton/01_point가공_어노테이션파일.xml')
xml.detectMultiScale(image, 1.3, 5)


import numpy as np
import cv2
from matplotlib import pyplot as plt
face_cascade = cv2.CascadeClassifier('./data/haar/haarcascade_frontface.xml')
eye_cascade = cv2.CascadeClassifier('./data/haar/haarcascade_eye.xml')
img = cv2.imread('./data/haar/img.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
